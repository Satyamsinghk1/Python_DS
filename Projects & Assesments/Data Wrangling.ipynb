# <b><u> Problem 2 :</u> Top 3 directors prefer which actors the most?

```python
# Your code here 
def find_directors(li):
  job_iterator = 0
  if type(li) != float:
    li = ast.literal_eval(li)
    while True:
      if li[job_iterator]['job'] != 'Director':
        job_iterator += 1
      else:
        return li[job_iterator]['name']

imdb_df['directors_name'] = imdb_df.apply(lambda x: find_directors(x['crew']),axis=1)
grouped_pop = imdb_df.groupby('directors_name')['popularity'].mean().sort_values(ascending = False).reset_index()
grouped_pop.head(3)

```
