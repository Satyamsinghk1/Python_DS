Given a list of n lists, write a function which returns the intersection of all the n lists. Each list can have any number of elements.

Example 1:

```
input = [[5, 5, 5, 8, 10], 
      [5, 5, 7, 8], 
      [5, 5, 8, 9, 11]]
```

```
output = [5, 5, 8]
```

Example 2:

```
input = [[5, 6], 
      [5, 5, 7, 8], 
      [5, 5, 8, 9]]
```


```
output = [5]
```

````python
lsts = [[5, 5], 
      [5, 5, 7, 8], 
      [5, 5, 8, 9]]
````

````python
def get_intersection_1(lsts):

  min_len = len(lsts[0])
  for l in lsts:
      if len(l)<=min_len:
          min_len = len(l)
          min_len_list = l

  d = {}
  for elem in min_len_list:
      if elem in d.keys():
          d[elem] = d[elem]+1
      else:
          d[elem] = 1

  for key in d.keys():
    for l in lsts:
      count = 0
      for item in l:
        if key==item:
          count = count + 1
      
      if d[key] > count:
        d[key] = count

  lst = []
  for key, value in d.items():
      for v in range(value):
          lst.append(key)
        
  return lst
````
