# <b><u>Problem 2 </u></b>

### Read face cream and facewash product sales data and show it using the bar chart. The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

```
# Your code here
facecream_facewash_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/AlmaBetter/Module1/Data Visualization/Assignments/dat/company_sales_data.csv')
facecream_facewash = facecream_facewash_df[['facecream' , 'facewash']]
facecream_facewash.plot(kind = 'bar')
plt.title('sales per month')
plt.ylabel('products')
plt.xlabel('months')
```

# <b><u>Problem 3</u></b>

### Display the number of units sold per month for each product using multiline plots. (i.e. separate plotline for each product ).

```
# Your code here
plt.plot(facecream_facewash_df[['facecream']], label = 'Facecream Sales Data', linewidth=3)
plt.plot(facecream_facewash_df[['facewash']], label = 'Facewash Sales Data', linewidth=3)
plt.plot(facecream_facewash_df[['toothpaste']], label = 'Toothpase Sales Data', linewidth=3)
plt.plot(facecream_facewash_df[['bathingsoap']], label = 'Bathing soap Sales Data', linewidth=3)
plt.plot(facecream_facewash_df[['shampoo']], label = 'Shampoo Sales Data', linewidth=3)
plt.plot(facecream_facewash_df[['moisturizer']], label = 'Moisturizer Sales Data', linewidth=3)

plt.title('The number of unit sold per month for each product')
plt.ylabel('products')
plt.xlabel('months')
plt.legend(loc = 'upper left')

```
